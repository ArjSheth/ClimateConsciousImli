```python
import pandas as pd
import matplotlib.pyplot as plt
import torch.nn as nn
import numpy as np
```

```python
no2 = pd.read_csv("/home/anwesha/IML_Project/data/NO2_manual_1213m.csv")
wind_u = pd.read_csv("/home/anwesha/IML_Project/data/WIND_U_manual_27929m.csv")
wind_v = pd.read_csv("/home/anwesha/IML_Project/data/WIND_V_manual_27929m.csv")
temp = pd.read_csv("/home/anwesha/IML_Project/data/TEMP_manual_27929m.csv")
co = pd.read_csv("/home/anwesha/IML_Project/data/CO_manual_1213m.csv")
```

```python

# creating merged dataframe with all the features
for df_ in [wind_u, wind_v, no2]:
    df_['timestamp'] = pd.to_datetime(df_['timestamp'])
    df_['latitude'] = df_['latitude'].round(3)
    df_['longitude'] = df_['longitude'].round(3)


wind = wind_u.merge(
    wind_v,
    on=['timestamp', 'latitude', 'longitude']
)

# ensure column names match what we use below
wind = wind.rename(columns={'WIND_U': 'wind_u', 'WIND_V': 'wind_v'})

wind['wind_speed'] = np.sqrt(
    wind['wind_u']**2 + wind['wind_v']**2
)

wind['wind_dir'] = np.arctan2(
    wind['wind_v'], wind['wind_u']
)

wind['wind_dir'] = np.arctan2(
    wind['wind_v'], wind['wind_u']
)
```
